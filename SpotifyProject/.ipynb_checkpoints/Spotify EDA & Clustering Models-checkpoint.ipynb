# Import Spotify Library

```python
import pandas as pd
df = pd.read_csv("spotifyLibrary.csv", encoding='utf-8')
```

```python
numeric_data = [x for x in df.columns if x not in ["key","mode","time_signature",
                                                   "type","uri","id","track_href","Unnamed: 0",
                                                   "analysis_url"]]
```

```python
numeric_data_df = df[numeric_data].set_index(df['id'])
numeric_data_df
```

# K-Means Clustering

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=15,random_state=42)
km = km.fit(scaledFeatures_df)
```

```python
df['kmeans'] = km.predict(scaledFeatures_df)
```

```python
df.groupby(['kmeans']).size().to_frame
#just looking at the different groups that appear after running kmeans is interesting
```

```python
#create and fit a range of models
km_list = list()

for clust in range(1,21):
    km = KMeans(n_clusters = clust, random_state=42)
    km = km.fit(scaledFeatures_df)
    km_list.append(pd.Series({
        'clusters':clust,
        'inertia':km.inertia_,
        'model':km
    }))
```

```python
plot_data = (pd.concat(km_list,axis=1)
            .T[['clusters','inertia']]
            .set_index('clusters'))

ax = plot_data.plot(marker='o',ls='-')
ax.set_xticks(range(0,21,2))
ax.set_xlim(0,21)
ax.set(xlabel='Cluster',ylabel='Inertia')
```

# Dimensionality Reduction

Want to explore reducing the number of columns in the data.

Using Scikit-learns pipeline function recreate the data pre-processing scheme above (transforming and scaling) using a pipeline. If you used a non-Scikeit learn function to transform the data (Log), checkkout the custom transformer class called FunctionTransformer.

```python
from sklearn.preprocessing import FunctionTransformer
from sklearn.pipeline import Pipeline

#custom log transformer
log_transformer = FunctionTransformer(np.log1p)

#Pipeline
estimators = [('log1p', log_transformer), ('minmaxscale', MinMaxScaler())]
pipeline = Pipeline(estimators)

#convert Original data using the newly created pipeline from above
PipeData = df.drop(['Unnamed: 0', 'mode','type','key',
                    'id', 'uri', 'track_href', 'analysis_url', 'time_signature', 'kmeans', 'agg'], axis=1)
data_pipe = pipeline.fit_transform(PipeData)
```
